```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Use serif font
plt.rcParams['font.family'] = 'Times'
plt.rcParams['text.usetex'] = True

# Generate a DataFrame with two numeric columns
np.random.seed(0)
n_samples = 100
X = np.random.rand(n_samples, 1) * 10  # Feature column
y = 2.5 * X.squeeze() + np.random.randn(n_samples) * 2  # Target column with some noise

df = pd.DataFrame({'X': X.squeeze(), 'y': y})

# Apply linear regression
model = LinearRegression()
model.fit(X, y)
y_pred = model.predict(X)

# Plot the result
plt.scatter(df['X'], df['y'], color='blue', label='Data points')
plt.plot(df['X'], y_pred, color='red', label='Linear fit')
plt.xlabel('X')
plt.ylabel('y')

# Title with LaTeX equation of the model
slope = model.coef_[0]
intercept = model.intercept_
equation = f'$y = {slope:.2f}x + {intercept:.2f}$'
plt.title(f'Linear Regression\n{equation}', fontsize=14)
plt.legend()
```

```python
from paretoset import paretoset

fig, ax = plt.subplots(nrows=1, ncols=1)
plt.scatter(df['X'], df['y'], color='blue')
mask = paretoset(df[["X", "y"]], sense=["min", "max"])
df[mask].plot.scatter(x='X', y='y', ax=ax, c='red')
```
